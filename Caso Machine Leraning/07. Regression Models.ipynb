```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_theme(style="white")

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error

%matplotlib inline
```

#### Regression model example

```python
df = pd.read_csv('data/ad_revenue.csv', index_col=0)

plt.figure(figsize=(12, 6))
plt.title('Sales performance related to the TV Ads')
plt.xlabel('TV Ads')
plt.ylabel('Sales')
sns.regplot(x='TV', y='sales', data=df, fit_reg=True, ci=None)
plt.show()
```

---

#### Spending for different age group

```python
# Import the dataset
df = pd.read_csv('data/age_spend.csv')

# Prepare independent and dependent variables
X = df[['age']]
y = df['spend']

# Split test & train datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101)
```

#### Linear Regression Model

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

lr_predictions = lr_model.predict(X_test)
print('Linear regression RMSE: ' + str(mean_squared_error(lr_predictions, y_test)**0.5))
```

#### Prepare validation dataset

```python
ages = pd.DataFrame({'age':range(18, 70)})
```

#### Linear regression result

```python
plt.figure(figsize=(12,8))
plt.scatter(X_train.age.tolist(), y_train.tolist())

plt.plot(ages, lr_model.predict(ages), color='r', label="Linear Regression")

plt.legend()
plt.xlabel("age")
plt.ylabel("spend")
plt.show()
```

#### Decision Tree

```python
# Create two decision trees. one with depth 2, another with 5
tree_d2 = DecisionTreeRegressor(max_depth=2)
tree_d2.fit(X_train, y_train)

tree_d5 = DecisionTreeRegressor(max_depth=5)
tree_d5.fit(X_train, y_train)

# Decission tree results in comparison with Linear regression results
lr_predictions = lr_model.predict(X_test)
print('Linear regression RMSE: ' + str(mean_squared_error(lr_predictions, y_test)**0.5))

tree_d2_predictions = tree_d2.predict(X_test)
print('Decision tree with max depth of 2 RMSE: ' + str(mean_squared_error(tree_d2_predictions, y_test)**0.5))

tree_d5_predictions = tree_d5.predict(X_test)
print('Decision tree with max depth of 5 RMSE: ' + str(mean_squared_error(tree_d5_predictions, y_test)**0.5))
```

#### Linear regression result with Decision tree results

```python
plt.figure(figsize=(12,8))
plt.scatter(X_train.age.tolist(), y_train.tolist())

plt.plot(ages, lr_model.predict(ages), color='r', label="Linear Regression")
plt.plot(ages, tree_d2.predict(ages),  color='g', label="Tree with max depth 2")
plt.plot(ages, tree_d5.predict(ages),  color='k', label="Tree with max depth 5")

plt.legend()
plt.xlabel("age")
plt.ylabel("spend")
plt.show()
```
